```python
# 📦 Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 📂 Step 2: Load Data
df = pd.read_csv("mobile_usage_behavioral_analysis.csv")

# 📊 Step 3: Clean and Inspect
df = df.dropna().drop_duplicates()
print(df.info())

# 👤 Step 4: User Demographics
df_demographics = df[['User_ID', 'Age', 'Gender']]
print(df_demographics.head())

# 📈 Step 5: Total App Usage
df_app_usage = df[['User_ID', 'Total_App_Usage_Hours']]
print(df_app_usage.head())

# ⏰ Step 6: Daily Screen Time
df_screen_time = df[['User_ID', 'Daily_Screen_Time_Hours']]
print(df_screen_time.head())

# 📱 Step 7: Number of Apps Used
df_apps_used = df[['User_ID', 'Number_of_Apps_Used']]
print(df_apps_used.head())

# 🌐 Step 8: Social Media Usage
df_social = df[['User_ID', 'Social_Media_Usage_Hours']]
print(df_social.head())

# 🎮 Step 9: Gaming Usage
df_gaming = df[['User_ID', 'Gaming_App_Usage_Hours']]
print(df_gaming.head())

# 🧰 Step 10: Productivity Usage
df_productivity = df[['User_ID', 'Productivity_App_Usage_Hours']]
print(df_productivity.head())

# 🗂️ Step 11: Combine Categories with Sample App Names
df['Top_Social_Media_App'] = 'Instagram'
df['Top_Productivity_App'] = 'Google Docs'
df['Top_Gaming_App'] = 'PUBG'

# Display sample
df[['User_ID', 'Top_Social_Media_App', 'Top_Productivity_App', 'Top_Gaming_App']].head()
```

```python
# Step 1: Find the category with the highest usage for each user
def get_most_used_category(row):
    usage = {
        'Social Media': row['Social_Media_Usage_Hours'],
        'Gaming': row['Gaming_App_Usage_Hours'],
        'Productivity': row['Productivity_App_Usage_Hours']
    }
    return max(usage, key=usage.get)

df['Most_Used_App_Category'] = df.apply(get_most_used_category, axis=1)

# Step 2: Map categories to example apps
category_to_app = {
    'Social Media': 'Instagram',
    'Gaming': 'PUBG',
    'Productivity': 'Google Docs'
}
df['Most_Used_App_Name'] = df['Most_Used_App_Category'].map(category_to_app)

# Step 3: Display sample
df[['User_ID', 'Most_Used_App_Category', 'Most_Used_App_Name']].head()
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(x='Most_Used_App_Name', data=df, palette='cool')
plt.title("Most Used App (Estimated by Usage Category)")
plt.xlabel("App Name")
plt.ylabel("Number of Users")
plt.tight_layout()
plt.show()
```
